```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
import json
import plotly.io as pil
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
from plotly.validators.scatter.marker import SymbolValidator

import plotly.io as pio
pio.renderers.default = 'vscode'
```

```python
url_apple= 'https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv'
aapl = pd.read_csv(url_apple)
aapl
```

```python
fig = px.line(aapl, x = 'Date', y='AAPL.Close', range_x=['2016-01-01', '2017-03-31'])
fig.show()
```

```python
fig = px.line(aapl, x = 'Date', y='AAPL.Close')
fig.update_xaxes(rangeslider_visible = True)
fig.show()
```

```python
fig = go.Figure(
    data = [
        go.Candlestick (
            x = aapl['Date'],
            open = aapl['AAPL.Open'],
            high = aapl['AAPL.High'],
            low = aapl['AAPL.Low'],
            close = aapl['AAPL.Close'],
        )
    ]
)
fig.show()
```

```python
gapminder = px.data.gapminder()
fig = px.area(gapminder, x = 'year', y ='pop', color = 'continent', line_group='country')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x = [1,2,3,4,5],
    y = [6,7,8,9,3],
    fill= 'tozeroy'
)
)
fig.add_trace(go.Scatter(
    x = [1,2,3,4,5],
    y = [3,9,4,12,3],
    fill= 'tozeroy'
)
)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x = [1,2,3,4,5],
    y = [6,7,8,9,3],
    fill= 'tozeroy',
    mode = 'none'
)
)
fig.add_trace(go.Scatter(
    x = [1,2,3,4,5],
    y = [3,9,4,12,3],
    fill= 'tozeroy',
    mode = 'none'
)
)
fig.show()
```
