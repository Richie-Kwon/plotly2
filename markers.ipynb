```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
import json
import plotly.io as pil
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
from plotly.validators.scatter.marker import SymbolValidator

import plotly.io as pio
pio.renderers.default = 'jupyterlab'
```

```python
iris = px.data.iris()
fig = px.scatter(iris, x = 'petal_width', y = 'petal_length',
                 color = 'species',
                size = 'sepal_length',
                hover_data=['sepal_width'])

fig.update_traces(
    marker = dict(size =12, line=dict(width=2, color='navy')),
    selector = dict(mode ='markers')

)
fig.show()       
```

```python
x1 = np.random.uniform(low=1, high=5, size=(200,))
y1 = np.random.uniform(low=1, high=5, size=(200,))
x2 = np.random.uniform(low=5, high=10, size=(200,))
y2 = np.random.uniform(low=1, high=5, size=(200,))

fig = go.Figure()
fig.add_trace(
go.Scatter(
    mode ='markers',
    x=x1,
    y=y1,
    opacity = 0.5,
    marker = dict( color = 'red',
                  size=10,
                 line =dict(color='darkred', width=2)),
    name = 'Opacity 0.5'
    )
)

fig.add_trace(
go.Scatter(
    mode ='markers',
    x=x2,
    y=y2,
    opacity = 1,
    marker = dict( color = 'limegreen',
                  size=10,
                 line =dict(color='green', width=2)),
    name = 'Opacity 1'
    )
)
fig.show()
```

```python
gapminder = px.data.gapminder()

fig = px.scatter(
    gapminder, x = 'pop', y = 'lifeExp',
    color = np.log10(gapminder['pop']),
    hover_name = 'country', log_x = True

)

fig.update_layout (
    coloraxis_colorbar = dict(
    title = 'Population',
    tickvals = [6,7,8,9],
    ticktext = ['1M', '10M', '100M', '1B']
    )

)

fig.show()
```
