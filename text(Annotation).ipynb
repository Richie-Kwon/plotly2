```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
import json
import plotly.io as pil
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
from plotly.validators.scatter.marker import SymbolValidator

import plotly.io as pio
pio.renderers.default = 'jupyterlab'
```

```python
gapminder = px.data.gapminder()

gapminder_asia_2007 = gapminder.query("year ==2007 and continent =='Asia'")
```

```python
fig = px.scatter(gapminder_asia_2007,
                x = 'gdpPercap',
                y = 'lifeExp',
                text = 'country',
                log_x = True,
                size_max =60                 
                )
fig.update_traces(textposition = 'top center')
fig.update_layout(height = 800)
fig.show()
```

```python
 fig = px.bar(gapminder_asia_2007,
             y = 'pop',
             x = 'country',
              text = 'pop',                     
             )
fig.update_traces(
    texttemplate = '%{text:.2s}', textposition = 'outside'
)
fig.update_layout()
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = [0,1,2,3,4,5,6,7,8],
                         y = [0,1,5,6,3,5,3,5,6]))
fig.add_trace(go.Scatter(x = [0,1,2,3,4,5,6,7,8],
                         y = [0,4,5,7,3,1,3,9,6]))
fig.add_annotation(x=2, y=4, text = 'Annotation1')
fig.add_annotation(x=4, y=4, text = 'Annotation2')
fig.update_annotations(
    dict(xref = 'x', yref='y', showarrow = True, arrowhead =7, ax=0, ay=-40)
)
fig.update_layout(showlegend = False)
fig.show()
```

```python
fig.add_annotation(x = 5, y =7, xref='x', yref='y',
                  text='another one', showarrow =True,
                  font =dict(family='Courier New, monospace',size=16,color='blue'),
                  align = 'center',
                  arrowhead=2,
                  arrowsize=1,
                  arrowcolor='orange',
                  ax = 20, ay =-40,
                  bordercolor = 'red',
                  borderwidth =2,
                  borderpad =4,
                  bgcolor = 'yellow',
                  opacity = 0.8)
fig.update_layout(showlegend = True)
fig.show()
```

```python
fig = px.line(gapminder, x ='year', y ='lifeExp', color='country')
fig.update_traces(mode = 'markers+lines')
fig.show()
```

```python
fig.update_traces(mode = 'markers+lines', hovertemplate=None)
fig.update_layout(hovermode = 'x')
fig.show()
```

```python
fig.update_traces(mode = 'markers+lines')
fig.update_layout(hoverlabel = dict(
    bgcolor = 'white',
    font_size=14,
    font_family="Courier New"
))
fig.show()
```

```python
fig = px.scatter(gapminder, x = 'gdpPercap', y ='lifeExp', log_x = True,
                 hover_name = 'country', hover_data=['continent', 'pop']
)
fig.show()
```

```python
fig.update_traces(hovertemplate = 'GDP: %{x} <br>Life Expectancy: %{y}')
fig.show()
```

```python
fig.update_traces(hovertemplate=None, selector={'name':'ASIA'})
fig.show()
```

```python
fig = px.line(gapminder, x ='year', y ='lifeExp', color ='country')
fig.update_traces(mode = 'markers+lines')
fig.update_xaxes(showspikes=True)
fig.update_yaxes(showspikes=True)
fig.show()
```

```python
fig.update_xaxes(showspikes=True, spikecolor='green', spikesnap = 'cursor', spikemode='across')
fig.update_yaxes(showspikes=True, spikecolor='orange', spikethickness=2)
fig.update_layout(spikedistance =1000, hoverdistance=100)
fig.show()
```
